```python
#Read all numeric columns in to Univ1
Univ1 = Univ.iloc[:,1:]
```

```python
Univ1
```

```python
cols = Univ1.columns
```

```python
#Standardisation function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Univ_df = pd.DataFrame(scaler.fit_transform(Univ1),columns = cols)
scaled_Univ_df
#scaler.fit_transform(Univ1)
```

```python
#build 3 clustsers using Kmeans cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=0) #Specify 3 clusters
clusters_new.fit(scaled_Univ_df)
```

```python
#print the cluster labels
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
#Assign clusters to the univ data set
Univ['clusterid_new'] = clusters_new.labels_
```

```python
Univ
```

```python
Univ.sort_values(by = "clusterid_new")
```

```python
#use groupby() to find aggregated (mean) values in each cluster
Univ.iloc[:,1:].groupby("clusterid_new").mean()
```

Obsevations

1.cluster 2 appears to be the top rated universities cluster as the cut off score,top10,SFRatio parameter mean values are highest
2.cluster 1 appears to occupy the middle level rated universities
3.cluster 0 comes as the lower level rated universities

```python
Univ[Univ['clusterid_new']==0]
```

```python
wcss = []
for i in range(1,20):
    
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_Univ_df)
    #kmeans.fit(Univ1)
    wcss.append(kmeans.inertia_)
print(wcss)
plt.plot(range(1,20), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

###observations:

1.from the above graph k = 3, 4 which indicates the ellow joint that is the rate of change of slope dcreases
