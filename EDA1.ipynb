```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data_clean.csv")
print(data)
```

```python
data.info()
```

```python
#dataframe atrributes
print(type(data))
print(data.shape)
print(data.size)
```

```python
#drop duplicate column(temp c) and unamed column 
data1 = data.drop(['Unnamed: 0',"Temp C"], axis=1)
data1
```

```python
#convert the month column data type to  float data type
data['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
#print all duplicated rows
data1[data1.duplicated(keep = False)]
```

```python
#drop duplicated rows
data1.drop_duplicates(keep='first', inplace = True)
data1
```

```python
#rename the columns
#CHANGE COLUMN NAME(RENAME THE COLUMNS)
data1.rename({'Solar.R': 'Solar'}, axis=1, inplace = True)
data1
```

```python
#imput the missing values
#display the data1 info()
data1.info()
```

```python
#display data1 missing values count in each column using isnull().sum()
data1.isnull().sum()
```

```python
#viusalize dat1 missing values using heat map
cols = data1.columns
colors = ['black', 'blue']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colors),cbar = True)
```

```python
#find the mean and median values of each
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
#replace the ozone missing values with median value
data1["Ozone"] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
#find the mode values of categorcial column(weather)
print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
#impute missing values(replace nan with mode etc.) of "weather" using mh=fillnan()
data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
### method 2
#using mu =/- 3*sigma limits(standard deviation method)
data1["Ozone"].describe()
```

```python
mu = data1["Ozone"].describe()[1]
sigma = data1["Ozone"].describe()[2]

for x in data1["Ozone"]:
    if ((x < (mu - 3*sigma)) or (x > (mu + 3*sigma))):
        print(x)
```

### observations
1 it is observed that only two outliers are idetified
2 in box plot method more no.of ouliers are identified
3 this is becaause the assumption of normallyis not satsifed in this column

```python
#Quantile-Quantile plot for detection of outliers
import scipy.stats as stats

# create q-q plot
plt.figure(figsize=(8,6))
stats.probplot(data1["Ozone"], dist="norm", plot=plt)
plt.title("Q-Q Plot for outlier Detection", fontsize=14)
plt.xlabel("Theoretical Quantiles", fontsize=12)
```

observations fro  Q-Q plot
1 the data does not follow normal distribution as the data points are deviationg sugnuficantly away from the red line
2 the data shows a right_skewed distribution and possible outliers
